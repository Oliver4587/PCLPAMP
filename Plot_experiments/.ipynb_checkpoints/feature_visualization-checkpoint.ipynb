```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
def pca_visualize(features, labels):
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(features)
    
    plt.figure(figsize=(10, 7))
    for label in set(labels):
        indices = [i for i, l in enumerate(labels) if l == label]
        if label == 0:
            print_label ="非抗菌肽"
        else:
            print_label = "抗菌肽"
        plt.scatter(principal_components[indices, 0], principal_components[indices, 1], label=print_label)
    plt.legend()
    plt.title('PCA 可视化特征表示')
    plt.xlabel('主成分 1')
    plt.ylabel('主成分 2')
    plt.show()

def tsne_visualize(features, labels):
    tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
    tsne_results = tsne.fit_transform(features)
    
    plt.figure(figsize=(10, 7))
    for label in set(labels):
        indices = [i for i, l in enumerate(labels) if l == label]
        
        if label == 0:
            print_label ="非抗菌肽"
        else:
            print_label = "抗菌肽"
        
        plt.scatter(tsne_results[indices, 0], tsne_results[indices, 1], label=print_label)
    plt.legend()
    plt.title('t-SNE Visualization')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

def pca_visualize_comparison(feature1, feature2, feature3, labels):
    # 定义PCA降维函数
    def pca_transform(features):
        pca = PCA(n_components=2)
        return pca.fit_transform(features)

    # 执行PCA
    pc1 = pca_transform(feature1)
    pc2 = pca_transform(feature2)
    pc3 = pca_transform(feature3)

    # 创建图形
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    datasets = [(pc1, "(a) 未训练模型提取的特征"), 
                (pc2, "(b) 提示调优方法训练的模型提取的特征"), 
                (pc3, "(c) 提示调优方法和对比学习方法结合训练的模型提取的特征")]

    for ax, (pc, title) in zip(axes, datasets):
        for label in set(labels):
            indices = [i for i, l in enumerate(labels) if l == label]
            print_label = "非抗菌肽" if label == 0 else "抗菌肽"
            ax.scatter(pc[indices, 0], pc[indices, 1], label=print_label, s=20)
        ax.set_title(title)
        ax.set_xlabel('主成分 1')
        ax.set_ylabel('主成分 2')
        ax.legend()
        ax.grid(True)

#     plt.suptitle('PCA 可视化特征表示对比')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('../Figures/INDEP/PCA_Feature_Comparison.png', bbox_inches='tight', dpi=300)
    plt.show()

# 示例调用
# pca_visualize_comparison(feature1, feature2, feature3, labels)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 7.5
plt.rcParams['axes.unicode_minus'] = False

def pca_visualize_comparison(feature1, feature2, feature3, labels):
    def pca_transform(features):
        pca = PCA(n_components=2)
        return pca.fit_transform(features)

    pc1 = pca_transform(feature1)
    pc2 = pca_transform(feature2)
    pc3 = pca_transform(feature3)

    fig, axes = plt.subplots(1, 3, figsize=(10, 3))
    datasets = [(pc1, "未微调模型提取的特征"), 
                (pc2, "提示调优方法训练的模型提取的特征"), 
                (pc3, "提示调优方法和对比学习方法结合训练的模型提取的特征")]

    markers = {0: '*', 1: 'o'}

    for idx, (ax, (pc, title)) in enumerate(zip(axes, datasets)):
        # 先绘制抗菌肽（label为1）
        for label in [1, 0]:
            indices = [i for i, l in enumerate(labels) if l == label]
            print_label = "抗菌肽" if label == 1 else "非抗菌肽"
            gray_level = 0.8 if label == 1 else 0.2
            ax.scatter(pc[indices, 0], pc[indices, 1], label=print_label, s=12, marker=markers[label],
                       facecolors=str(gray_level), linewidths=0.5, edgecolors='black')
        ax.set_title(title, fontsize=7.5)
        ax.set_xlabel('主成分 1', fontsize=7.5)
        ax.set_ylabel('主成分 2', fontsize=7.5)
        ax.grid(True)
        
        # 在左上角添加标签
        ax.text(-0.10, 1.07, f'({chr(97 + idx)})', transform=ax.transAxes, fontsize=7.5, va='top')

    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.025), fontsize=7.5, ncol=2)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('../Figures/INDEP/gray/PCA_Feature_Comparison_WithStar.svg', bbox_inches='tight', dpi=300)
    plt.show()

# 示例调用
# pca_visualize_comparison(feature1, feature2, feature3, labels)
```
